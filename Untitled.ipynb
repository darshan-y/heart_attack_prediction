# Train Test Split

```python
from sklearn.model_selection import train_test_split
parameters = DS.drop("target",axis=1)
output = DS["target"]
X_train,X_test,Y_train,Y_test = train_test_split(parameters,output,test_size=0.20,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model1 = LogisticRegression(C=0.472, class_weight='balanced', dual=False,
                   fit_intercept=True, intercept_scaling=1, l1_ratio=None,
                   max_iter=1000, multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=123, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
model1.fit(X_train,Y_train)
prediction1 = model1.predict(X_test)

accu = round(accuracy_score(prediction1,Y_test)*100,2)
print("Logistic Regression accuracy : " + str(accu) + "%")
```

### Support Vector Machines (SVM) 

```python
from sklearn import svm
model2 = svm.SVC(kernel='linear')
model2.fit(X_train, Y_train)
prediction2 = model2.predict(X_test)

accu = round(accuracy_score(prediction2,Y_test)*100,2)
print("SVM accuracy : " + str(accu) + "%")
```

### K-Nearest Neighbour 

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=7)
model3.fit(X_train,Y_train)
prediction3=model3.predict(X_test)

accu = round(accuracy_score(prediction3,Y_test)*100,2)
print("K-Nearest Neighbour accuracy : " + str(accu) + "%")
```

### XGBoost 

```python
import xgboost as xgb
model4 = xgb.XGBClassifier(objective="binary:logistic", random_state=42)
model4.fit(X_train, Y_train)
prediction4 = model4.predict(X_test)

accu = round(accuracy_score(prediction4,Y_test)*100,2)
print("XGBoost accuracy : " + str(accu) + "%")
```

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
model5 = DecisionTreeClassifier()
model5.fit(X_train, Y_train)
prediction5 = model5.predict(X_test)
accu = round(accuracy_score(prediction5,Y_test)*100,2)
print("DecisionTree accuracy : " + str(accu) + "%")
```

```python
from sklearn.tree import plot_tree

inputs=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
output = np.array('target')

plt.figure(figsize=(20, 15))
plot_tree(model5, feature_names=inputs,class_names=str(np.unique(output)), filled=True)
plt.show()
```

### Random forest 

```python
from sklearn.ensemble import RandomForestClassifier
model6 = RandomForestClassifier()
model6.fit(X_train,Y_train)
prediction6 = model6.predict(X_test)
accu = round(accuracy_score(prediction6,Y_test)*100,2)
print("Random Forest accuracy : " + str(accu) + "%")
```
